## Quickstart with Transpile

```python
from qiskit import transpile
qc = QuantumCircuit(3)
qc.y(0)
for t in range(2): 
    qc.cx(0, t + 1)
qc.draw("mpl", style = "iqx")
```

```python
from qiskit.test.mock import FakeSantiago
santiago = FakeSantiago()
santiago.configuration().basis_gates
```

```python
t_qc = transpile(qc, santiago)
t_qc.draw("mpl", style = "iqx")
```

```python
santiago.configuration().coupling_map
```

```python
t_qc = transpile(qc, santiago, initial_layout=
                                        [1,0,2])
t_qc.draw("mpl")
```

## Transpiler Passes

### The PassManager

```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(3)
qc.h(0)
qc.cx(0, 2)
qc.cx(2, 1)
qc.draw("mpl")
```

```python
from qiskit.transpiler import PassManager,CouplingMap
from qiskit.transpiler.passes import BasicSwap

coupling_map = CouplingMap([[0,1], [1,2]])
basic_swap_pass = BasicSwap(coupling_map)
pm = PassManager(basic_swap_pass)
```

```python
routed_qc = pm.run(qc)
routed_qc.draw("mpl")
```
